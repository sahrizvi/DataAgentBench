```python
import json
from pathlib import Path

def load_json_lines(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            return [json.loads(line.strip()) for line in f if line.strip()]
    except Exception as e:
        print(f"❌ Error loading {file_path}: {e}")
        return []

def save_json_lines(data, output_path):
    try:
        with open(output_path, "w", encoding="utf-8") as f:
            for record in data:
                f.write(json.dumps(record) + "\n")
        print(f"✅ Saved {len(data)} records to {output_path}")
    except Exception as e:
        print(f"❌ Error saving {output_path}: {e}")

def relabel_gmap_ids(meta_data, review_data):
    # Step 1: Create consistent mapping from original gmap_id → new_id (e.g., gmap_0)
    id_map = {}
    for i, entry in enumerate(meta_data):
        original_id = entry.get("gmap_id")
        new_id = f"gmap_{i}"
        id_map[original_id] = new_id
        entry["gmap_id"] = new_id  # Replace in meta_data

    # Step 2: Replace in review_data
    replaced = 0
    for r in review_data:
        original = r.get("gmap_id")
        if original in id_map:
            r["gmap_id"] = id_map[original]
            replaced += 1

    print(f"✅ Replaced {len(id_map)} unique gmap_ids in meta_data")
    print(f"✅ Replaced {replaced} gmap_ids in review_data")
    return meta_data, review_data

if __name__ == "__main__":
    # === File paths ===
    meta_path = "dataset/googlelocal/light_meta_LLM_tt.json"
    review_path = "dataset/googlelocal/light_review_LLM_tt.json"
    new_meta_path = "dataset/googlelocal/light_meta_LLM_tt.json"
    new_review_path = "dataset/googlelocal/light_review_LLM_tt.json"

    # === Load original data ===
    meta_data = load_json_lines(meta_path)
    review_data = load_json_lines(review_path)

    # === Replace gmap_id ===
    meta_new, review_new = relabel_gmap_ids(meta_data, review_data)

    # === Save to new files ===
    save_json_lines(meta_new, new_meta_path)
    save_json_lines(review_new, new_review_path)
```
