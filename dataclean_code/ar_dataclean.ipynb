```python
import json
import random
from tqdm import tqdm
from openai import AzureOpenAI

# ====== Azure OpenAI 配置 ======
client = AzureOpenAI(
    api_key="609ced4d971240b8a08f7fb0e6d846ea",
    api_version="2024-08-01-preview",
    azure_endpoint="https://promptdelta-nc.openai.azure.com",  # 不要加 /v1
)

deployment_name = "gpt-4o-mini"  # 在 Azure Portal 的 Deployments 中查看

# ====== 文件路径 ======
input_path = "dataset/googlelocal/light_review_with_address.json"
output_path = "dataset/googlelocal/light_review_naturally_enriched.json"

# ====== 地址提取 ======
def extract_location_from_address(address):
    """从地址中提取 City / State / City+State 三种格式之一，随机选一种"""
    if not address or ',' not in address:
        return None
    parts = address.strip().split(',')
    if len(parts) >= 3:
        city = parts[-2].strip()
        state_zip = parts[-1].strip()
        full = f"{city}, {state_zip}"
        return random.choice([city, full])
    return None

# ====== 调用 LLM 改写或生成评论文本 ======
def enrich_text_with_location(original_text, location_phrase):
    if not location_phrase:
        return original_text or ""

    if original_text:
        prompt = (
            f"The customer review is:\n\"{original_text}\"\n"
            f"Write a new version by keeping the original review unchanged, "
            f"and add a short natural sentence about the location '{location_phrase}' either before or after the review. "
            "Do not rephrase the review itself. The final result should contain exactly the original review and one extra sentence about the location."
        )
    else:
        prompt = (
            f"Write a one-sentence natural customer review for a business located in '{location_phrase}'. "
            "Make it sound helpful or positive, and include the location in the sentence."
        )

    try:
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {"role": "system", "content": "You are a helpful assistant that rewrites and enriches customer reviews."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
            max_tokens=80
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"❌ Error enriching review: {e}")
        return original_text or ""

# ====== 主处理流程 ======
with open(input_path, "r", encoding="utf-8") as fin, \
     open(output_path, "w", encoding="utf-8") as fout:

    for line in tqdm(fin):
        entry = json.loads(line)

        address = entry.get("address", "")
        text_raw = entry.get("text", "")
        original_text = text_raw.strip() if isinstance(text_raw, str) else ""

        location_phrase = extract_location_from_address(address)
        new_text = enrich_text_with_location(original_text, location_phrase)
        entry["text"] = new_text

        fout.write(json.dumps(entry, ensure_ascii=False) + "\n")

print(f"✅ Location-enriched reviews saved to: {output_path}")
```

```python
import pandas as pd
import random
from datetime import datetime

# ====== 路径配置 ======
input_path = "dataset/googlelocal/light_review_naturally_enriched.json"
output_path = "dataset/googlelocal/light_review_LLM.json"

# ====== 时间格式化函数 ======
def format_timestamp_randomly(ts_ms):
    try:
        dt = datetime.fromtimestamp(ts_ms / 1000)
        formats = [
            dt.strftime("%Y-%m-%d %H:%M:%S"),           # ISO 风格
            dt.strftime("%B %d, %Y at %I:%M %p"),       # 美式自然语言
            dt.strftime("%d %b %Y, %H:%M"),             # 英式简洁风格
        ]
        return random.choice(formats)
    except:
        return ts_ms  # 如果时间格式错误，保留原始值

# ====== 加载数据 ======
df = pd.read_json(input_path, lines=True)

# ====== 删除 address 字段（如果有） ======
if "address" in df.columns:
    df.drop(columns=["address"], inplace=True)
    print("✅ 'address' 列已删除")
else:
    print("⚠️ 'address' 列不存在，无需删除")

# ====== 转换时间字段 ======
if "time" in df.columns:
    df["time"] = df["time"].apply(lambda ts: format_timestamp_randomly(ts) if pd.notnull(ts) else ts)
    print("✅ 'time' 字段已转换为自然语言格式")
else:
    print("⚠️ 'time' 字段不存在，未做处理")

# ====== 保存结果 ======
df.to_json(output_path, orient="records", lines=True)
print(f"✅ 文件已保存到：{output_path}")
```

```python
import json
from tqdm import tqdm
from openai import AzureOpenAI

# ====== Azure OpenAI 配置 ======
client = AzureOpenAI(
    api_key="609ced4d971240b8a08f7fb0e6d846ea",
    api_version="2024-08-01-preview",
    azure_endpoint="https://promptdelta-nc.openai.azure.com",  # 不要加 /v1
)

deployment_name = "gpt-4o-mini"  # 在 Azure Portal 的 Deployments 中查看

# ====== 文件路径 ======
INPUT_PATH = "dataset/googlelocal/light_meta.json"
OUTPUT_PATH = "dataset/googlelocal/light_meta_withdescip.json"

# ====== 生成描述（包含 category + location） ======
def generate_category_description(category_list, address):
    category = category_list[0] if isinstance(category_list, list) and category_list else "local business"

    # 提取地址中的城市 + 州 ZIP（如 "Los Angeles, CA 90023"）
    location = ""
    if isinstance(address, str) and ',' in address:
        parts = address.strip().split(',')
        if len(parts) >= 3:
            city = parts[-2].strip()
            state_zip = parts[-1].strip()
            location = f"{city}, {state_zip}"

    # 构造提示词
    prompt = (
        f"Write a natural and concise one-sentence English description for a business that is a {category} "
        f"located in {location}. "
        f"The sentence must include the phrase '{category}' and mention the location '{location}'. "
        "Do not mention the business name. Use a helpful, informative tone."
    )

    try:
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {"role": "system", "content": "You are a helpful assistant who writes business descriptions."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
            max_tokens=60
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"❌ Error generating for category '{category}': {e}")
        return None

# ====== 主循环（可改为前10行为 debug） ======
with open(INPUT_PATH, "r", encoding="utf-8") as fin, \
     open(OUTPUT_PATH, "w", encoding="utf-8") as fout:

    for i, line in enumerate(tqdm(fin)):
        # 只处理前10行可用于 debug
        # if i >= 10:
        #     break

        entry = json.loads(line)
        category_list = entry.get("category", [])
        address = entry.get("address", "")
        old_desc = entry.get("description") or ""

        if category_list:
            new_sentence = generate_category_description(category_list, address)

            if new_sentence:
                if old_desc and not old_desc.endswith('.'):
                    old_desc += "."
                entry["description"] = f"{old_desc.strip()} {new_sentence}"
            else:
                entry["description"] = old_desc  # fallback if generation failed

        fout.write(json.dumps(entry, ensure_ascii=False) + "\n")

print(f"✅ Descriptions updated and saved to: {OUTPUT_PATH}")
```

```python
import json

INPUT_PATH = "dataset/googlelocal/light_meta_withdescip.json"
OUTPUT_PATH = "dataset/googlelocal/light_meta_LLM.json"

fields_to_remove = ["category", "address", "latitude", "longitude"]

with open(INPUT_PATH, "r", encoding="utf-8") as fin, \
     open(OUTPUT_PATH, "w", encoding="utf-8") as fout:

    for line in fin:
        entry = json.loads(line)

        for field in fields_to_remove:
            entry.pop(field, None)  # 安全删除：即使字段不存在也不会报错

        fout.write(json.dumps(entry, ensure_ascii=False) + "\n")

print(f"✅ Removed {fields_to_remove} and saved to: {OUTPUT_PATH}")
```
