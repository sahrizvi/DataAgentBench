```python
import pandas as pd

# 读取 jsonl
df = pd.read_json("spider2-snow-0713.jsonl", lines=True)
pd.set_option('display.max_colwidth', None)
# 指定要筛选的 db_id
target_db = "PATENTS"

# 按 db_id 筛选
df_filtered = (
    df[df["db_id"] == target_db]
    .sort_values(by="instance_id")  # 按 instance_id 排序
    .reset_index(drop=True)         # 重置索引
)


# 查看结果
df_filtered

```
